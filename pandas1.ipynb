<h5>Data Frame</h5>

```python
#From Dictionary of Lists
data = {
    'Name': ['vijay', 'kiran', 'narayana','ajay'],
    'Age': [21,23,25,27],
    'City': ['krishna', 'prakasam', 'srikakulam','kadapa']
}
pd.DataFrame(data)
```

```python
#From list of Dictionaries
data = [
    {'Name': 'vijay', 'Age': 21, 'City': 'krishna'},
    {'Name': 'kiran', 'Age': 23, 'City': 'prakasam'},
    {'Name': 'narayana', 'Age': 25, 'City': 'srikakulam'},
    {'Name': 'ajay', 'Age': 27, 'City': 'kadapa'}
]
pd.DataFrame(data)

```

```python
#from Dictionary of Series
data = {
    'Name': pd.Series(['vijay', 'kiran', 'narayana','ajay']),
    'Age': pd.Series([21,23,25,27]),
    'City': pd.Series(['krishna', 'srikakulam', 'prakasam','kadapa'])
}
pd.DataFrame(data)

```

```python
#From list of Lists
data = [
    [21, 'kiran', 'ongole'],
    [23, 'ajay', 'kadapa'],
    [25, 'gowtham', ' chittor'],
    [27,'narayana','guntur']
]
pd.DataFrame(data, columns=['Age', 'Name', 'City'])

```
